```python
import matplotlib.pyplot as plt
#Gráfico de línea:
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
plt.show()
```

```python
#Gráfico de barras:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.bar(x, y)
plt.show()
```

```python
#Gráfico de dispersión:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.scatter(x, y)
plt.show()
```

Crea una clase llamada Grafico que tenga un método linea que dibuje un gráfico de línea. El método debe aceptar dos listas: x y y, que contienen los datos para el eje x y el eje y, respectivamente.

```python
import matplotlib.pyplot as plt

class Grafico:
    def linea(self, x, y):
        plt.plot(x, y)
        plt.show()

# Ejemplo de uso
g = Grafico()
g.linea([1, 2, 3, 4], [1, 4, 2, 3])
```

Crea una clase llamada DiagramaDeBarras que tenga un método dibujar que dibuje un gráfico de barras. El método debe aceptar dos listas: x y y, que contienen los datos para el eje x y el eje y, respectivamente.

```python
import matplotlib.pyplot as plt

class DiagramaDeBarras:
    def dibujar(self, x, y):
        plt.bar(x, y)
        plt.show()

# Ejemplo de uso
d = DiagramaDeBarras()
d.dibujar([1, 2, 3, 4], [1, 4, 2, 3])
```

Crea una clase llamada DiagramaDeDispersión que tenga un método dibujar que dibuje un gráfico de dispersión. El método debe aceptar dos listas: x y y, que contienen los datos para el eje x y el eje y, respectivamente.

```python
import matplotlib.pyplot as plt

class DiagramaDeDispersión:
    def dibujar(self, x, y):
        plt.scatter(x, y)
        plt.show()

# Ejemplo de uso
d = DiagramaDeDispersión()
d.dibujar([1, 2, 3, 4], [1, 4, 2, 3])
```
