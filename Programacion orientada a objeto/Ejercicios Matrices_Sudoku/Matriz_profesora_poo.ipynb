Una clase MatrizCuadrada que hereda de la anterior y reescribe el siguiente método público
- inicializar: dada la dimensión de la matriz, pone a cero todos sus elementos.
- redimensionar: cambia las dimensiones de la matriz por la especificada, poniendo además todos sus elementos a cero.  
y añade los siguientes métodos públicos:
- obtener_matriz_adjunta: devuelve la matriz adjunta de la posición especificada.
- calcular_determinante: devuelve el determinate de la matriz.



```python
class Matriz:
    def __init__(self, n_filas, n_columnas):
        self.__elementos = []
        for e in range(n_filas):
            self.__elementos.append([0]*n_columnas)
        #print(self.__elementos)

    def __str__(self):
        representacion = ''
        for fila in self.__elementos:
            representacion += str(fila) + '\n'
        return representacion

    def obtener_numero_filas(self):
        return len(self.__elementos)

    def obtener_numero_columnas(self):
        return len(self.__elementos[0])

    def obtener_elemento(self, fila, columna):
        return self.__elementos[fila][columna]

    def cambiar_elemento(self, posicion, nuevo_elemento):
        self.__elementos[posicion[0]][posicion[1]] = nuevo_elemento

    # Como parámetro va a ser una lista con el valor de fila y columna
    def redimensionar(self, dimension):
        # dimension[0], dimension[1] hace referencia a los índices
        # m.redimensionar([4,3])
        self.__init__(dimension[0], dimension[1])

    def mostrar(self):
        return self.__elementos


m1 = Matriz(3, 3)

m1.mostrar()




```

```python
class MatrizConOperaciones(Matriz):
    def sumar(self, Mat):
        """
        Realiza la suma de la matriz actual con otra matriz Mat.

        Verifica que ambas matrices tengan la misma dimensión y luego suma los elementos
        correspondientes de las dos matrices y los asigna a la matriz actual.
        """
        if (self.obtener_numero_columnas() == Mat.obtener_numero_columnas()) and (
            self.obtener_numero_filas() == Mat.obtener_numero_filas()):
            for fila in range(self.obtener_numero_filas()):
                for columna in range(self.obtener_numero_columnas()):
                    # Selecciono la fila y columna de las dos matrices y luego
                    # sumo el elemento 1 de la matriz con el elemento 2
                    self.cambiar_elemento((fila, columna), self.obtener_elemento(
                        fila, columna) + Mat.obtener_elemento(fila, columna))
        else:
            print('Las matrices no tienen la misma dimensión y por tanto no pueden sumarse')

    def multiplicar_por_escalar(self, n):
        """
        Multiplica todos los elementos de la matriz por un escalar n dado.

        Itera sobre cada elemento de la matriz y lo multiplica por n.
        """
        for fila in range(self.obtener_numero_filas()):
            for columna in range(self.obtener_numero_columnas()):
                self.cambiar_elemento((fila, columna), self.obtener_elemento(
                    fila, columna) * n)

    def transponer(self):
        """
        Calcula la matriz transpuesta intercambiando filas por columnas.

        Crea una nueva matriz transpuesta y asigna los elementos correspondientes a partir de la matriz actual.
        Luego, redimensiona la matriz actual para que coincida con la dimensión de la matriz transpuesta.
        """
        # Crea la lista con las listas vacías
        trapuesta = []
        columnas = self.obtener_numero_columnas()
        filas = self.obtener_numero_filas()
        for columna in range(columnas):
            valores_traspuestos = []
            for fila in range(filas):
                valores_traspuestos.append(self.obtener_elemento(fila, columna))
            trapuesta.append(valores_traspuestos)

        # Lleva la matriz a ceros
        self.redimensionar((columnas, filas))
        # Recorremos la matriz por sus índices
        for fila in range(len(trapuesta)):
            for columna in range(len(trapuesta[0])):
                self.cambiar_elemento((fila, columna), trapuesta[fila][columna])

    def multiplicar(self, matriz):
        """
        Realiza la multiplicación de la matriz actual con otra matriz dada.

        Verifica que el número de columnas de la matriz actual sea igual al número de filas
        de la matriz dada para que la multiplicación sea posible.
        Luego, crea una nueva matriz y calcula los elementos multiplicando los elementos
        correspondientes de las dos matrices y sumándolos.
        Finalmente, asigna los elementos de la matriz resultante a la matriz actual y redimensiona
        la matriz actual con la dimensión de la matriz resultante.
        """
        if self.obtener_numero_columnas() != matriz.obtener_numero_filas():
            raise ValueError("Error! No se pueden multiplicar estas matrices, la dimensión no es la adecuada")
        else:
            # Crea una nueva matriz con ceros (verifica  que tenga las mismas filas columnasd)
            matriz_nueva = MatrizConOperaciones(self.obtener_numero_filas(), matriz.obtener_numero_columnas())

            # Cambia fila por columna y viceversa para sacar el resultado en una nueva matriz
            # porque es más fácil recorrer filas que las columnas de las filas
            matriz.transponer()

            for fila in range(matriz_nueva.obtener_numero_filas()):
                for columna in range(matriz_nueva.obtener_numero_columnas()):
                    suma = 0
                    for it in range(self.obtener_numero_columnas()):
                        suma += self.obtener_elemento(fila, it) * matriz.obtener_elemento(columna, it)
                    matriz_nueva.cambiar_elemento((fila, columna), suma)

        self.redimensionar((matriz_nueva.obtener_numero_filas(), matriz_nueva.obtener_numero_columnas()))
        for fila in range(matriz_nueva.obtener_numero_filas()):
            for columna in range(matriz_nueva.obtener_numero_columnas()):
                self.cambiar_elemento((fila, columna), matriz_nueva.obtener_elemento(fila, columna))
```

```python
m1 = MatrizConOperaciones(3, 3)
m2 = MatrizConOperaciones(3, 3)

m1.cambiar_elemento([0, 0], 1)
m1.cambiar_elemento([0, 1], 2)
m1.cambiar_elemento([0, 2], 3)
m1.cambiar_elemento([1, 0], 3)
m1.cambiar_elemento([1, 1], 2)
m1.cambiar_elemento([1, 2], 1)
m1.cambiar_elemento([2, 0], 0)
m1.cambiar_elemento([2, 1], 1)
m1.cambiar_elemento([2, 2], 2)

m2.cambiar_elemento([0, 0], 2)
m2.cambiar_elemento([0, 1], 2)
m2.cambiar_elemento([0, 2], 3)
m2.cambiar_elemento([1, 0], 3)
m2.cambiar_elemento([1, 1], 2)
m2.cambiar_elemento([1, 2], 4)
m2.cambiar_elemento([2, 0], 1)
m2.cambiar_elemento([2, 1], 1)
m2.cambiar_elemento([2, 2], 1)

m1.multiplicar(m2)

#print(m1)
#m1.transponer()
#print(m1)
```
