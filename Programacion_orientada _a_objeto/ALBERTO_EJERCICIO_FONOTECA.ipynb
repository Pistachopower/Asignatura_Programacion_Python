- obtener_intérpretes: devuelve los intérpretes de la canción.
- obtener_duración: devuelve una tupla con los minutos y segundos que dura la canción.
- editar_canción: cambia los datos de la canción por los especificados.

```python
class Cancion:
    def __init__(self,T,I,min,seg):
        self.__titulo = T
        self.__interprete = I
        self.__minutos = min
        self.__segundos = seg
    def __str__(self):
        return 'Titulo: '+self.__titulo+ '\n'+' Interprete: '+self.__interprete+'\n'+' Duracion: '+str(self.__minutos)+'\n'+' Segundos: '+str(self.__segundos)

    def obtener_titulo(self):
        return self.__titulo
    
    def obtener_interprete(self):
        return self.__interprete
    
    def obtener_duracion(self):
        return (self.__minutos,self.__segundos)

    def editar_cancion(self,T1,I1,min1,seg1):
        self.__titulo = T1
        self.__interprete = I1
        self.__minutos = min1
        self.__segundos = seg1
```

```python
class Album:
    def __init__(self,T,A):
        self.__Titulo = T
        self.__Artista = A
        self.__Canciones = []

    def __str__(self):
        canciones = ''
        for cancion in self.__Canciones:
            canciones += str(cancion)
        return 'Titulo:'+self.__Titulo+'\n' +'Artista: '+ self.__Artista +'\n'+ 'Canciones: ' + canciones

    def obtener_titulo(self):
        return self.__Titulo

    def obtener_artistas(self):
        return self.__Artista
    
    def editar_album(self,T1,A1):
        self.__Titulo = T1
        self.__Artista = A1

    def incluir_cancion(self,C):
        self.__Canciones.append(C)

    def eliminar_cancion(self,posicion):
        self.__Canciones.pop(posicion)

    def obtener_cancion(self,posicion):
        return self.__Canciones[posicion]

    def obtener_duracion(self):
        (min,seg)=(0,0)
        for cancion in self.__Canciones:
            a,b = cancion.obtener_duracion()
            min +=a
            seg += b
        while seg>=60:
            min+=1
            seg -= 60
        return (min,seg)
```

```python
class Fonoteca:
        def __init__(self,P):
                self.__Propietario = P
                self.__Albumes = []

        def __str__(self):
                cadena = ''
                for album in self.__Albumes:
                        cadena += str(album)
                return 'Propietario:' + self.__Propietario +'Albumes: ' + cadena

        def incluir_album(self,A):
                self.__Albumes.append(A)

        def eliminar_album(self,posicion):
                self.__Albumes.pop(posicion)

        def obtener_album(self,posicion):
                return self.__Albumes[posicion]
    
```
