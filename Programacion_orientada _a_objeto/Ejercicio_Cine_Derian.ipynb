```python
class Pelicula:
    def __init__(self, tit, tim, act, dire):
        self.__titulo = tit
        self.__tiempo = tim
        self.__actor = act
        self.__director = dire
        

    def __str__(self):
        return 'Titulo: ' + '\n' + str(
            self.__titulo) + '\n' + 'Tiempo: ' + str(
            self.__tiempo) + '\n' + 'Actor: ' + str(
            self.__actor) + '\n' + 'Director: ' + str(
            self.__director) + '\n' 

    def obtener_titulo (self):
        return self.__titulo

    def obtener_duracion(self):
        return self.__tiempo

    def obtener_actor(self):
        return self.__actor
    
    def editar_pelicula(self, titulo, tiempo, actor, director):
        self.__titulo = titulo
        self.__tiempo = tiempo
        self.__actor = actor
        self.__director =director

```

```python
class Cine:
    def __init__(self, cartelera):
        self.cartelera= cartelera
        self.peliculas = []

    def __str__(self):
        datos = '\n' + 'Datos de peliculas: ' + '\n' 

        for pelicula in self.peliculas:
            datos += str(pelicula)
        return str(datos)

    def incluir_pelicula(self, pelicula):
        self.peliculas.append(pelicula)

    def obtener_peliculas(self, indice):
        print(self.peliculas[indice-1]) 



    #obtener el titulo
```

```python
cine= Cine('cartelera')

cine.incluir_pelicula(pelicula)
cine.incluir_pelicula(pelicula2)

#print(cine)

#cine.obtener_peliculas(1)
```

```python

class Taquilla (Cine):
    def __init__(self, taq, pre_entr, ent_vend):
        self.taquillas= taq
        self.precio_entrada= pre_entr
        self.entradas_vendidas= ent_vend

    def __str__(self):
        return '\n' + 'Taquillas: ' + str(self.taquillas) + '\n' + 'Precio entrada: ' + str(
        self.precio_entrada) + '\n' + 'Entradas vendidas: ' + str(self.entradas_vendidas)

    def obtener_entradas_vendidas(self):
        return self.entradas_vendidas
    
    def obtener_precio_entradas(self):
        return self.precio_entrada
    
    def obtener_taquillas(self):
        return  self.taquillas
    
    def editar_precio_entradas(self, precio):
        self.precio_entrada= precio

    def aplicar_descuento_entradas(self, descuento):
        self.precio_entrada= str(round((self.precio_entrada /100) * descuento,2))
        
    
```
