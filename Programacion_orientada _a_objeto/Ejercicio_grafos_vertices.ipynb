```python
class Grafo:
#v= 4
#e=[(0,1),(1,2),(2,3)]
#
    def __init__(self, v, e):
        self.__vertices = v
        self.__ejes= e

    def __str__(self):
        return 'Total vertices: ' + str(
            self.__vertices) + '\n' +'Ejes del grafo: \n' + str(
                self.__ejes) + "\n " 

    def obtener_orden(self):
        return self.__ejes

    def obtener_tamano(self):
        return len(self.__ejes)

    def son_adyacentes(self, adyacencias):
        if adyacencias in self.__ejes:
                respuesta= 'Son adyacentes'
        else:
             respuesta= 'No son adyacentes'
        return respuesta
                
        
        

grafo= Grafo(4,[(0,1),(1,2),(2,3)])

print(grafo.son_adyacentes((1,0)))
```

```python
class GrafoColoreado(Grafo):
    def __init__(self, colors=None):
        self.__colores= []

    #def __str__(self):
    #    return str(self.__colores)
    
    def obtener_color_vertice(self, posicion):
        return self.__colores[posicion]
    
    def colorear_vertice(self, color):
        self.__colores.append(color)

    def comprobar_coloracion(self):
        #variable que recorre los indices de self.__colores [especificamente el valor del indice]
        posicion= 1
        for color in self.__colores:
            if color == self.__colores[posicion]:
                break
            posicion+=1

            #cuando posicion sea >= a la longitud de self.__colores termina la funcion 
            #se hace para que no de el error fuera de rango
            if posicion >= len(self.__colores):
                break
                
        


pintar= GrafoColoreado([])

#pintar= GrafoColoreado([])

#pintar.obtener_color_vertice(2)

pintar.colorear_vertice(1)

pintar.colorear_vertice(2)

pintar.colorear_vertice(3)

pintar.colorear_vertice(4)

pintar.comprobar_coloracion()

#print(pintar)

```
