Un sudoku es un pasatiempo lógico que se realiza en un tablero 9 x 9, dividido en 9 cuadrados 3 x 3a. El objetivo consiste en completar el tablero de forma que cada fila, cada columna y cada cuadrado contenga los números del 1 al 9 sin que falte ni se repita ninguno. Se proporcionan al comienzo los valores de algunas de las casillas del tablero.

Una clase Sudoku con los siguientes atributos públicos:
- Tablero: guarda una matriz con los valores de las casillas del tablero, como números enteros entre 0 y 9 (el cero corresponde a una casilla con valor desconocido).   
Y métodos públicos:   
- inicializar: inicializa el tablero con los valores especificados en una lista de tuplas (i,j,num), donde i y j son las coordenadas de una casilla y num es el valor que tiene esa casilla.

- representar: devuelve una cadena que representa el tablero.

- rellenar_casilla: establece el valor de la casilla especificada, siempre que este no se conociera previamente.   

```python
class Sudoku:
    def __init__(self, valores_ini):
        self.tablero = []
        for i in range(9):
            self.tablero.append([0]*9)  # Crear una matriz vacía de 9x9 (tablero)
        for tp in valores_ini:
            i = tp[0]  # Obtener la fila del valor inicial
            j = tp[1]  # Obtener la columna del valor inicial
            num = tp[2]  # Obtener el número del valor inicial
            self.tablero[i][j] = num  # Asignar el número al tablero en la posición correspondiente

    def __str__(self):
        representacion = ''
        for fila in self.tablero:
            representacion += str(fila) + '\n'  # Convertir la fila en una cadena y agregarla a la representación
        return representacion

    def rellenar_casilla(self, i, j, num):
        if self.tablero[i][j] == 0:  # Verificar si la casilla está vacía
            self.tablero[i][j] = num  # Asignar el número a la casilla
        else:
            # Lanzar una excepción si la casilla ya contiene un número
            raise Exception('La casilla está ocupada por el numero {}'.format(
                self.tablero[i][j]))
```

```python
S1 = Sudoku([(0, 1, 7), (0, 2, 2), (1, 0, 3), (1, 2, 4), (2, 0, 8), (2, 1, 1), (2, 2, 9)])
print(S1)

#S1.rellenar_casilla(0, 0, 5)  # Rellenar una casilla del Sudoku
#print(S1)
```

Una clase SudokuConAyuda que herede de la clase anterior y añada los siguientes métodos públicos:
- comprobar_casilla: determina si es posible rellenar la casilla especificada con el valor indicado.
- pedir_ayuda: devuelve una lista de tuplas (i,j,num), donde i y j son las coordenadas de las casillas que solo pueden contener un único valor, y num es ese valor.

```python
class SudokuConAyuda(Sudoku):
    def comprobar_fila(self, i, num):
        return num not in self.tablero[i]  # Verificar si el número no está presente en la fila

    def comprobar_columna(self, j, num):
        for i in range(9):
            if self.tablero[i][j] == num:  # Verificar si el número está presente en alguna casilla de la columna
                return False
        return True

    def comprobar_casilla(self, i, j, num):
        if self.tablero[i][j] != 0:  # Verificar si la casilla ya está ocupada
            return False
        else:
            return self.comprobar_fila(i, num) and self.comprobar_columna(j, 
                    num) and self.comprobar_cuadrado(i, j, num)


    def comprobar_cuadrado(self, i, j, num):
        fila = i - i % 3  # Calcular el índice de la fila del cuadrado en el que se encuentra la casilla
        columna = j - j % 3  # Calcular el índice de la columna del cuadrado en el que se encuentra la casilla
        for ii in range(fila, fila + 3):
            for jj in range(columna, columna + 3):
                if self.tablero[ii][jj] == num:  # Verificar si el número está presente en alguna casilla del cuadrado
                    return False
        return True
    
    def pedir_ayuda(self):
        casillas_unicas = []
        for i in range(9):
            for j in range(9):
                if self.tablero[i][j] == 0:  # Verificar si la casilla está vacía
                    valores_posibles = []
                    for num in range(1, 10):
                        if self.comprobar_casilla(i, j, num):  # Verificar si el número es válido para la casilla
                            valores_posibles.append(num)
                    if len(valores_posibles) == 1:  # Verificar si solo hay una opción válida para la casilla
                        casillas_unicas.append((i, j, valores_posibles[0]))  # Agregar la casilla a la lista de casillas únicas
        return casillas_unicas  # Devolver la lista de casillas únicas
```
