El juego del buscaminas es un juego lógico de un solo jugador cuyo objetivo es despejar un campo de minas sin detonar ninguna de ellas. Cuando el juego comienza se le presenta al jugador un tablero cuadrado con todas las casillas ocultas. El jugador va descubriendo esas casillas una a una; si la casilla elegida ocultaba una mina, el jugador ha perdido; si la casilla elegida no ocultaba una mina, entonces se le muestra al jugador el número de minas ocultas en las ocho (o cinco o tres, según la posición) casillas adyacentes. El jugador también tiene la posibilidad de marcar con una bandera cualquier casilla que piense que oculta una mina, o de desmarcar una casilla marcada si piensa que no oculta una mina.

Una clase CampoDeMinas con los siguientes atributos públicos
dimensión: guarda la dimensión del tablero de juego como un número entero positivo.
tablero: guarda el tablero de juego como una matriz cuadrada cuyos elementos son tuplas de la forma (estado, contenido), donde estado puede ser una de las tres cadenas 'O' (la casilla está oculta), 'B' (la casilla está marcada con una bandera) o 'D' (la casilla está descubierta) y contenido es un valor lógico indicando si la casilla contiene o no una mina.
y métodos públicos
inicializar: establece el tablero inicial del juego, dados su dimensión y la lista de casillas que contienen una mina. Téngase en cuenta que inicialmente todas las casillas del tablero están ocultas.

estado_casilla: devuelve el estado de la casilla especificada. 

valor_casilla: si la casilla especificada contiene una mina, entonces devuelve -1; en caso contrario, devuelve el número de minas que hay en las casillas adyacentes.

representar: devuelve una representación del tablero, donde cada fila se representa en una línea diferente, las casilla ocultas se representan por la letra O, las casillas marcadas se representan por la letra B, las casillas descubiertas que contienen una mina se representan por la letra M y las que no la contienen por el número de minas de las casillas adyacentes.

```python

class Campo_de_minas:
#inicializar: establece el tablero inicial 
#del juego, dados su dimensión y la lista de 
#casillas que contienen una mina. Téngase en cuenta 
#que inicialmente todas las casillas del tablero están ocultas.

              #dim= entero casillas= lista de tuplas con enteros de la 
              #posicion que van las minas
   def __init__(self, dim, casillas_minas):
       self.dimension = dim
       self.tablero = []

       #genera el tablero
       for i in range(dim):
           fila = [('O', False)] * dim
           self.tablero.append(fila)
           
        #del tablero coloca las minas
       for casilla in casillas_minas:
           #agrego las posiciones de las filas columnas de las tuplas
           f = casilla[0]
           c = casilla[1]
           #me posiciono en el tablero y pongo las minas 
           self.tablero[f][c] = ('O', True)

#estado_casilla: devuelve el estado de la casilla especificada. 
   def estado_casilla(self, fila, col):
       #selecciona fila, columna e indice de la tupla 
       return self.tablero[fila][col][0]

#valor_casilla: si la casilla especificada contiene una mina, entonces 
#devuelve -1; en caso contrario, devuelve el número de minas que hay 
# en las casillas adyacentes.
   def valor_casilla(self, fila, col):
       #selecciona el contenido (true o false) del tablero
       mina = self.tablero[fila][col][1]

       #si contenido de mina es == True devuelve -1
       if mina:
           return -1
       
       #devuelve el número de minas que hay en las casillas adyacentes
       else:
           minas_alrededor = 0
           #sube a la fila o se mantiene en la fila 0
           fila_sup = max(fila-1,0)

           #baja a la fila o se mantiene en la fila si no hay mas fila
           fila_inf = min(fila+1, self.dimension-1)

           #se desplaza a la izquierda de la columna o se 
           # mantiene en la columna sino tiene otra 
           col_izq = max(col-1, 0)

           #se desplaza a la derecha de la columna o se 
           # mantiene en la columna sino tiene otra 
           col_der = min(col + 1, self.dimension-1)
           
           #recorremos con los rangos definidos
           for f in range(fila_sup, fila_inf + 1):
               for c in range(col_izq, col_der + 1):
                   #con los valores de los rangos de fila columna 
                   # recorremos el tablero por indices
                   #por cada iteracion y modificamos las casillas
                   mina = self.tablero[f][c][1]

                   #si mina es True suma 1 a variable minas_alrededor
                   if mina:
                       minas_alrededor += 1
           return minas_alrededor
   
#representar: devuelve una representación del tablero, donde 
#cada fila se representa en una línea diferente, las casilla 
#ocultas se representan por la letra O, las casillas marcadas 
#se representan por la letra B, las casillas descubiertas que 
#contienen una mina se representan por la letra M y las que no 
#la contienen por el número de minas de las casillas adyacentes.
  
   def __str__(self):
       #aquí solo se muestra el tablero oculto
       representacion = ''
       for f in range(self.dimension):
           for c in range(self.dimension):
               #obtiene '0', 'B','D' 
               estado = self.estado_casilla(f,c)
               if estado == 'O':
                   representacion += 'O '
               elif estado == 'B':
                   representacion += 'B '
               elif estado == 'D':
                   #obtiene true, false o -1 si tiene mina
                   valor = self.valor_casilla(f,c)
                   if valor == -1:
                       representacion += 'M '
                   else:
                       representacion += str(valor) + ' '
           representacion += '\n'
       return representacion

#tablero= Campo_de_minas(4, [(0,0),(0,1),(0,3),(1,0),(1,1),(1,3), 
#                            (2,0), (2,1),(2,2),(2,3)])
#print(tablero.estado_casilla(0,0))
#tablero.valor_casilla(0,2)
#tablero.valor_casilla(1,2)
#print(tablero)


```

Una clase Buscaminas que herede de la clase anterior y añada los métodos públicos
marcar_casilla: marca con una bandera la casilla especificada; en caso de que esta no este oculta, no se puede marcar.

desmarcar_casilla: desmarca con una bandera la casilla especificada; en caso de que esta no esté marcada, no se puede desmarcar.

descubrir_casilla: descubre la casilla especificada, en caso de que esta esté oculta o marcada con una bandera. Si ni la casilla ni ninguna de las adyacentes contienen una mina (es decir, el valor de la casilla especificada es cero), entonces se descubren automáticamente todas las casillas adyacentes que aún no lo estén.


```python
class Buscaminas (Campo_de_minas):
#RECUERDA: todo lo que hagas se modifica en el tableto interno, 
#pero al momento de imprimir se muestra el estado del tablero, no el
# contenido (estado,contenido) 
  
   def marcar_casilla (self, fila, col):
       #selecciono estado '0'
       estado = self.estado_casilla(fila,col)
       
       if estado == 'O':
           #selecciono el valor logico true o false (si hay o no hay mina)
           mina = self.tablero[fila][col][1]
           # cambio el '0' por 'B' junto con el valor de mina
           self.tablero[fila][col] = ('B', mina)
       else:
           raise Exception ('Esa casilla no se puede marcar')
   
   #debe estar marcada la casilla primero   
   def desmarcar_casilla (self, fila, col):
       estado = self.estado_casilla(fila,col)

       if estado == 'B':
           #selecciono el contenido true o false (que hay minas)
           mina = self.tablero[fila][col][1]
           #y lo cambio por el valor que está en  ()
           #recuerda que como es una tupla debe modificarse toda la tupla
           self.tablero[fila][col] = ('O', mina)
       else:
           raise Exception ('Esa casilla no se puede desmarcar')
      
   def descubrir_casilla (self, fila, col):
       estado = self.estado_casilla(fila, col)
       #se cumple cuando usamos el mismo indice
       if estado == 'D':
           raise Exception ('Esa casilla ya está descubierta')
       else:
            #selecciona el valor true o false si hay o no minas
           mina = self.tablero[fila][col][1]
           #cambiamos la tupla a descubierta 
           self.tablero[fila][col] = ['D', mina]
           
           #valor= dependiendo si hay mina devuelve (-1):M
           # o sino lo hay devuelve las minas adyacentes 
           # o en otro lado
    
            #sino hay minas adyacentes busca en todo el tablero
           valor = self.valor_casilla(fila, col)
           if valor == 0:
               fila_sup = max(fila - 1, 0)
               fila_inf = min(fila + 1, self.dimension - 1)
               col_izq = max(col - 1, 0)
               col_der = min(col + 1, self.dimension - 1)
               for f in range(fila_sup, fila_inf + 1):
                   for c in range(col_izq, col_der + 1):
                       estado = self.estado_casilla(f,c)
                       if estado != 'D':
                           self.descubrir_casilla(f, c)

#definimos dimension del tablero y ponemos las minas
C = Buscaminas(4, [(0,0)])

#descubrimos minas
C.descubrir_casilla(3,0)
print(C)
```
