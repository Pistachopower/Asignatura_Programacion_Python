```python
class Alumno:
    def __init__(self, nombre):
        self.nombre = nombre
        self.notas = []
    
    def __str__(self):
        return '\nNombre: ' + str(self.nombre) + '\n' + 'Notas: ' + str(self.notas)

    def poner_notas(self, add_nota):
        self.notas.append(add_nota)

alumno1= Alumno('Juan')

alumno2= Alumno('Pedro')

alumno1.poner_notas(10)
alumno1.poner_notas(5)
alumno1.poner_notas(6)

alumno2.poner_notas(4)
alumno2.poner_notas(3)
alumno2.poner_notas(5)

#alumno2.poner_notas([9.5,6.6,7,7])

print(alumno1)
print(alumno2)
```

```python
class Asignatura:
    def __init__(self, nombre_asig):
        self.nombre_asig = nombre_asig
        self.alumnos = []
    
    def __str__(self):
        valor= 'Nombre asignatura: ' + '\n' + str(
            self.nombre_asig) + '\n' + 'Alumnos: '

        datos_alumnos= ''
        
        for datos in self.alumnos:
            datos_alumnos+= str(datos)

        return str(valor) + str(datos_alumnos)


    def incluir_alumno(self, alumno):
        self.alumnos.append(alumno)

    def quitar_alumno(self, alumno):
        self.alumnos.pop(alumno)

asig= Asignatura('programacion')

asig.incluir_alumno(alumno1)
asig.incluir_alumno(alumno2)

#asig.quitar_alumno(1)

print(asig)

```

```python
class Evaluacion(Asignatura):
    def nota_final(self, nombre):
        for i in range((len(self.alumnos))):
            alumno= self.alumnos[i]
            if alumno.nombre == nombre:
                nota= sum(alumno.notas)
                break
        if nota > 10:
            nota= 10
        return nota

    def evaluar_asignatura(self):
        for alumno in self.alumnos:
            nombre= alumno.nombre
            nota= self.nota_final(nombre)
            print('Alumno: {} '.format(nombre))
            print('Nota final: {} '.format(nota))

e= Evaluacion('programacion')
#print(e)

e.incluir_alumno(alumno1)
e.incluir_alumno(alumno2)
#print(e)

#print(e)
#e.nota_final('Juan')
#e.nota_final('Pedro')

#e.evaluar_asignatura()
```
