```python
import math


class Punto:


   def __init__(self, x, y):
       self.coordX = x
       self.coordY = y


   def __str__(self):
       return 'X: ' + str(self.coordX) + '\n' + 'Y: ' + str(self.coordY)


   def sumar(self, P):
       self.coordX += P.coordX
       self.coordY += P.coordY
  
   def multiplicar_por_escalar(self, r):
       self.coordX *= r.coordX
       self.coordY *= r.coordY


   def distancia(self, P):
       return math.sqrt(
           (P.coordX - self.coordX) ** 2 +
           (P.coordY - self.coordY) ** 2
       )


A= Punto(1.3, 2.2)
B= Punto(2.3, 1.2)
C= Punto(1.3, 2.9)

#A.sumar(B)
#print(A)

#A.multiplicar_por_escalar(B)
A.distancia(B)
print(A)



 
```

```python
class Triangulo:
   def __init__(self, A, B, C):
       self.verticeA = A
       self.verticeB = B
       self.verticeC = C


   def __str__(self):
       return 'vertice A: ' + str(self.verticeA) + '\n' + 'vertice B:  ' + str(self.verticeB) + '\n' + 'verticeC: ' + str(self.verticeC) + '\n'

    #hace la union y suma de los vertices de los lados a,b,c
   def calcular_perimetro (self):
       AB =  self.verticeA.distancia(self.verticeB)
       AC = self.verticeA.distancia(self.verticeC)
       BC = self.verticeB.distancia(self.verticeC)
       return AB + AC + BC


   def calcular_area (self):
       #aplica los pasos de la formula
       #une los vertices del triangulo
       a = self.verticeA.distancia(self.verticeB)
       b = self.verticeB.distancia(self.verticeC)
       c = self.verticeA.distancia(self.verticeC)
       #saca el resultado del perimetro y lo divido entre 2
       p = self.calcular_perimetro() / 2
       #saca el resultado del area con la formula de Heron 
       return math.sqrt(p * (p - a) * (p - b) * (p-c))

triangulo= Triangulo(A,B,C)

triangulo.calcular_area()

#triangulo.calcular_perimetro()
```
