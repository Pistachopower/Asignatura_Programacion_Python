```python
class Calendario: 
    def __init__(self,a):
        self.__anio= a
        self.__mes= 1
        self.__dia= 1
        
        
    def __str__(self):
        representacion= 'Día: ' + str(self.__dia) + '\n' + 'Mes: ' + str(
            self.__mes) + '\n' + 'Año: ' + str(self.__anio)
        return representacion
    
    def obtener_dia(self):
         return self.__dia        
     
    def obtener_mes(self):
         return self.__mes
     
    def obtener_anio(self):
         return self.__anio       
     
    def avanzar_dia(self):
        self.__dia = self.__dia + 1
        if self.__dia > 32:
            self.__dia = - 1
            self.__mes = self.__mes + 1
            if self.__mes == 13:
                self.__mes= -1
                self.__anio = self.__anio +1
        return self.__dia
    
    def avanzar_mes(self):
        self.__mes = self.__mes + 1
        if self.__mes == 13:
            self.__mes = - 1
            self.__anio = self.__anio + 1
        return self.__mes    

    def avanzar_anio(self):
        self.__anio = self.__anio + 1
        return self.__anio
    
    def retroceder_dia(self):
        self.__dia= self.__dia - 1 
        if self.__dia == 0:
            self.__dia = 31
            self.__mes = self.__mes -1
            if self.__mes == 0:
                self.__mes= 12
                self.__anio = self.__anio - 1
        return self.__dia
    
    def retroceder_mes(self):
        self.__mes = self.__mes -1
        if self.__mes == 0:
            self.__mes = 12
            self.__anio = self.__anio -1
        return self.__mes
    
    def retroceder_anio(self):
        self.__anio = self.__anio - 1
        return self.__anio

```
