```python
class Cancion:
    def __init__(self, t, i, m, s):
        self.__titulo = t
        self.__interprete= i 
        self.__minutos= m
        self.__segundo= s

    def __str__(self):
        return '* Titulo: ' + str(self.__titulo) + '\n ' + 'Interprete:' +str(
            self.__interprete) + '\n' +'Duracion: ' + str(self.__minutos) + ':' + str(
                self.__segundo) + '\n'
        
    def obtener_titulo(self):
        return self.__titulo

    def obtener_interprete(self):
        return self.__interprete

    def obtener_duracion(self):
        return (self.__minutos,  self.__segundo)
    
    def editar_cancion(self, edit_t, edit_i, edit_m, edit_s):
        self.__titulo = edit_t
        self.__interprete= edit_i
        self.__minutos= edit_m
        self.__segundo= edit_s
```

```python
class Album:
    def __init__(self, t_a, a_a):
        self.__titulo_album= t_a
        self.__artista_album= a_a
        self.__canciones_album= []

        
    def __str__(self):
        representar= 'Titulo album: ' + str(self.__titulo_album) + '\n ' + 'Artista album:' +str(
            self.__artista_album) + '\n' +'canciones album: ' + '\n' 
        
        for cancion in self.__canciones_album:
            representar += str(cancion) 
        return representar
    
    def obtener_titulo_album(self):
        return self.__titulo_album
    
    def obtener_artista_album(self):
        return self.__artista_album
    
    #al cambiar se mantiene las canciones del album modificado -mirar-
    def editar_album(self, ed_titulo_album, ed_artista_album):
        self.__titulo_album= ed_titulo_album
        self.__artista_album= ed_artista_album

    def incluir_cancion_album(self, ad_cancion_lista_album):
        self.__canciones_album.append(ad_cancion_lista_album)

    def eliminar_cancion(self,posicion):
        self.__canciones_album.pop(posicion-1)


    def obtener_cancion(self, indice): 
        return self.__canciones_album[indice-1]

    def obtener_duracion(self):
        minuto=0
        segundo=0
        for cancion in self.__canciones_album:
            a,b= cancion.obtener_duracion()
            minuto= minuto + a
            segundo= segundo + b
        while segundo >= 60:
            minuto= minuto + 1
            segundo= segundo - 60
            
        return 'Minuto:', (minuto), '| segundos:', (segundo)
```

```python
class Fonoteca: 
    def __init__(self, pro):
        self.__propietario= pro
        self.__albumes= []
        
    def __str__(self):
        representar= '*Propietario fonoteca: '+ self.__propietario + '\n' + 'Albumes fonoteca: '+ '\n'
    
        for album in self.__albumes:
            representar += str(album)
        return representar


    def incluir_album(self, ad_album_fonoteca):
        self.__albumes.append(ad_album_fonoteca)

    def eliminar_album(self, posicion):
        self.__albumes.pop(posicion-1)

    def obtener_album(self, indice):
        return self.__albumes[indice-1]
```
